```python
import numpy as np
import csv
import pandas as pd
import os
import json
```

```python
# import the dataset into a dataframe
df = pd.read_csv('synthetic_hiring_bias_dataset.csv')
```

```python
# pull one of each for each category matching
categories = {
    'gender': ['Male', 'Female'],
    'race': ['White', 'Other'],
    'age_group': ['Young', 'Older', 'Middle'],
    #'was_disadvantaged': [0,1],     # should have been hired (qualified)
    'hired': [0,1]                  # was actually hired (result)
}
```

```python
result_rows = []

for gender in categories['gender']:
    for race in categories['race']:
        for age_group in categories['age_group']:
            # for was_disadvantaged in categories['was_disadvantaged']:
                for hired in categories['hired']:
                    matching = df[
                        (df['gender'] == gender) &
                        (df['race'] == race) &
                        (df['age_group'] == age_group) &
                        # (df['was_disadvantaged'] == was_disadvantaged) &
                        (df['hired'] == hired)
                    ]
                    if not matching.empty:
                        # randomly select one row from the matching entries
                        m = matching.sample(n=1)
                        result_rows.append(m.iloc[0])
                    else:
                        # print(f"No match entry for gender={gender}, race={race}, age_group={age_group}, was_disadvantaged={was_disadvantaged}, hired={hired}")
                        print(f"No match entry for gender={gender}, race={race}, age_group={age_group}, hired={hired}")

combo_entries = pd.DataFrame(result_rows).reset_index(drop=True)
```

```python
# show the resulting dataframe
print(f"Entries: {len(combo_entries)}")
print(combo_entries)
```

```python
# export to csv
combo_entries.to_csv('study_applicants.csv', index=False)
```
